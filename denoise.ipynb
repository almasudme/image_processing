```python
import numpy as np
import cv2
from os.path import basename,splitext
from matplotlib import pyplot as plt
width =18
height =12

input_image = 'a.jpg'
image_name=basename(input_image)
image_name=splitext(image_name)[0]
#print(image_name)
#color image in grayscale
img = cv2.imread(input_image,0)

plt.figure(figsize=(width, height))
plt.hist(img.ravel(),256,[0,256])
plt.show()
```

```python
image_mean = cv2.blur(img,(5,5))
plt.figure(figsize=(width, height))
plt.imshow(image_mean,'gray')
plt.show()
```

```python
image_median= cv2.medianBlur(img,3)
plt.figure(figsize=(width, height))
plt.imshow(image_median,'gray')
plt.show()
```

```python
image_median= cv2.medianBlur(img,5)
plt.figure(figsize=(width, height))
plt.imshow(image_median,'gray')
plt.show()
```

```python
image_gaussian=cv2.GaussianBlur(img,(5,5),0)

plt.figure(figsize=(width, height))
plt.imshow(image_gaussian,'gray')
plt.show()
```
